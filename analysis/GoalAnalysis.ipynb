```python
from json_to_dataframe import base_dataframe

df = base_dataframe()
```

```python
top_10_scorers = df.head(10)
top_10_scorers.plot.bar(x='name', y='goals_scored')
```

```python
teams = df.groupby('team')['goals_scored'].sum()
teams.plot.bar()
```

```python
nationalities = df.groupby('nationality')['goals_scored'].sum()
nationalities.sort_values(ascending=False, inplace=True)
top_10_nations = nationalities[:10]
top_10_nations.plot.pie()
```

```python
liverpool = df.query("team == 'Liverpool'")
print(liverpool)
liverpool.groupby('name')['goals_scored'].sum().plot(kind='pie', y='goals_scored')
```
